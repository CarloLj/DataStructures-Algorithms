# A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
km = KMeans(n_clusters=2)
yp = km.fit_predict(dataset[["x1", "x2"]])
yp
```

Add to your dataset a column with the assihned cluster to each data point

```python
dataset["NewColumn"] = yp
dataset
```

Print the number associated to each cluster

```python
dataset.NewColumn.unique()
```

Print the centroids

```python
km.cluster_centers_
```

Print the intertia metric

```python
km.inertia_
```

Plot a scatter plot of your data assigned to each cluster. Also plot the centroids

```python
df1 = dataset[dataset.NewColumn==0]
df2 = dataset[dataset.NewColumn==1]

plt.scatter(df1.x1, df1.x2, label='group x1')
plt.scatter(df2.x1, df2.x2, label='group x2')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='magenta', marker='*', label='Centroides', s=256)

plt.title('x1 and x2')
plt.xlabel('x2')
plt.ylabel('x1')
plt.legend()
plt.show()
```

# A4 Elbow plot

Compute the Elbow plot

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(dataset[['x1','x2']])
    sse.append(km.inertia_)
    
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```

### Questions

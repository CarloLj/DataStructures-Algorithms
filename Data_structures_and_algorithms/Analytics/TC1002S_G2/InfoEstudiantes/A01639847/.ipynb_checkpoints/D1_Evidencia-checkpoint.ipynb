# A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
yp = km.fit_predict(dataset[['x2','x3']])

# Print estimated cluster of each point in the dataser
yp
```

Add to your dataset a column with the assigned cluster to each data point

```python
# Add a new column to the dataset with the cluster information
dataset['Cluster_Type'] = yp

# Visualize the dataset
dataset
```

Print the number associated to each cluster

```python
dataset.Cluster_Type.unique()
```

Print the centroids

```python
# Center of each cluster
km.cluster_centers_
```

Print the intertia metric

```python
# Sum of squared error
km.inertia_
```

Plot a scatter plot of your data assigned to each cluster. Also plot the centroids

```python
df1 = dataset[dataset.Cluster_Type==0]
df2 = dataset[dataset.Cluster_Type==1]
df3 = dataset[dataset.Cluster_Type==2]

plt.scatter(df1.x2, df1.x3, Label='Estimated Group 1')
plt.scatter(df2.x2, df2.x3, Label='Estimated Group 2')
plt.scatter(df3.x2, df3.x3, Label='Estimated Group 2')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='magenta', marker='*', label='Centroides', s=256)

plt.title('x1 vs x2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

# A4 Elbow plot

Compute the Elbow plot

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(dataset1[['x1','x2']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```

The **Seaborn** provides visualisations tools that will allow to explore data from a graphical perspective.

```python
# Import the packages that we will be using
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset url
path = "datasets/iris/iris.csv"

# Load the dataset
df = pd.read_csv(path)
```

```python
sns.histplot(df["SepLen"])
plt.show()
sns.histplot(df["SepWd"])
plt.show()
sns.histplot(df["PtlLen"])
plt.show()
sns.histplot(df["PtlWd"])
plt.show()
```

```python
xdata = df.loc[:,["SepLen", "SepWd", "PtlLen", "PtlWd"]]

sns.boxplot(data = xdata, orient = "h")


plt.show()
```

```python
sns.boxplot(y=df["PtlWd"], x=df["Class"]).set_title("Boxplot de ancho de pétalo por tipo de flor")

plt.show()
```

```python
sns.boxplot(y=df["SepLen"], x=df["Class"]).set_title("Boxplot de largo de sepal por tipo de flor")

plt.show()
```

Tanto los pétalos como los sepal son considerablemente más largos que anchos. Los pétalos tienen una distribución más dispersa que los sepal, es decir, presentan una mayor desviación estándar, adicionalmente, sus medidas son menores. La clase virginica es la más grande y la setosa la más pequeña (con pocas excepciones). Por útltimo, la frecuencia del ancho del sepal aparenta tener una distribución binomial a diferencia del resto de variables sin patrón aparente.

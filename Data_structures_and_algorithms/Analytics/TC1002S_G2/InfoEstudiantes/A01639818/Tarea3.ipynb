```python
import pandas as pd

ruta = "datasets/iris/iris.csv"


iris = pd.read_csv(ruta)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris.head()
```

```python
sns.displot(iris["Sepal Length"])
plt.show()
```

```python
sns.displot(iris["Sepal Width"])
plt.show()
```

```python
sns.displot(iris["Petal Length"])
plt.show()
```

```python
sns.displot(iris["Petal Width"])
plt.show()
```

```python
sns.displot(iris["FlowerType"])
plt.show()
```

```python
g = sns.FacetGrid(iris,row="FlowerType")
g = g.map(plt.hist,"Sepal Length")

plt.show()
```

```python
data = sns.boxplot(iris["Sepal Length"])
data.set(title = "Diagrama de longitud de setal")
plt.show
```

```python
data = sns.boxplot(iris["Sepal Width"])
data.set(title = "Diagrama de ancho de sepal")
plt.show
```

```python
data = sns.boxplot(iris["Petal Length"])
data.set(title = "Diagrama de longitud del Petal")
plt.show
```

```python
data = sns.boxplot(iris["Petal Width"])
data.set(title = "Diagrama de ancho de Petal")
plt.show
```

```python
sns.boxplot(y=iris["Petal Width"], x=iris["FlowerType"]).set(title = "De ancho del petalo por el tipo de flor")

plt.show()
```

```python
sns.boxplot(y=iris["Sepal Length"], x=iris["FlowerType"]).set(title = "De longitud por el tipo de flor")

plt.show()
```

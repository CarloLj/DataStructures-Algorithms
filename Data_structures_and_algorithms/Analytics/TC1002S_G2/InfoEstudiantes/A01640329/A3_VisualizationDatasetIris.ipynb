```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ruta = "datasets/iris/iris.csv"
dataset = pd.read_csv(ruta)
```

# 1

```python
x = sns.histplot( dataset["setalL"])
x.set_ylabel("Frecuencia")
x.set_title("Histograma de la largo de seta")
plt.show()
```

```python
x = sns.histplot( dataset["setalW"])
x.set_ylabel("Frecuencia")
x.set_title("Histograma del ancho de seta")
plt.show()
```

```python
x = sns.histplot( dataset["petalL"])
x.set_ylabel("Frecuencia")
x.set_title("Histograma de la largo de petalo")
plt.show()
```

```python
x = sns.histplot( dataset["petalW"])
x.set_ylabel("Frecuencia")
x.set_title("Histograma del ancho de petalo")
plt.show()
```

```python
x = sns.histplot( dataset["Class"])
x.set_ylabel("Frecuencia")
x.set_title("Histograma de tipos de iris")
plt.show()
```

# 4

```python
sns.boxplot(x=dataset["petalW"], y=dataset["Class"]).set_title("Boxplot de ancho de petalo de cada flor")
plt.show()
```

```python
sns.boxplot(x=dataset["setalL"], y=dataset["Class"]).set_title("Boxplot de largo de seta de cada flor")
plt.show()
```

# 6
Observando la información de los histogramas y boxplots se puede ver que la mayoria de las flores tienen un largo de seta aproximado a 6cm y un ancho de seta de 3cm. También se logra observar que el largo y el ancho de los petalos tienden a ser pequeños, variando en un largo de 1cm y en un ancho menor a .5cm, sin embargo, también se observa la presencia de numerosas flores con un largo y ancho un poco más grande, variando entre 4.5 y 1.3 respectivamente, así dando un promedio de 3.7 de largo y 1.3 de ancho.

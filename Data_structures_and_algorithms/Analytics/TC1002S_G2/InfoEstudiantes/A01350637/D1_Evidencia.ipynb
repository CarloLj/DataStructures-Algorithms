# A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python

# Import sklearn KMeans
from sklearn.cluster import KMeans

# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
yp = km.fit_predict(df[['x1','x2']])

# Print estimated cluster of each point in the dataser
yp
```

Add to your dataset a column with the assihned cluster to each data point

```python
# Add a new column to the dataset with the cluster information
df['Cluster1'] = yp

df
```

Print the number associated to each cluster

```python
df.Cluster1.unique()
```

Print the centroids

```python
km.cluster_centers_
```

Print the intertia metric

```python
km.inertia_
```

Plot a scatter plot of your data assigned to each cluster. Also plot the centroids

```python

df1 = df[df.Cluster1==0]
df2 = df[df.Cluster1==1]
df3 = df[df.Cluster1==2]


plt.scatter(df1.x1, df1.x2, label='Grupo 1')
plt.scatter(df2.x1, df2.x2, label='Grupo 2')
plt.scatter(df3.x1, df3.x2, label='Grupo 3')


plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='green', marker='*', label='Centroides', s=256)

plt.title('x2 vs x1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

# A4 Elbow plot

Compute the Elbow plot

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(df[['x1','x2']])
    sse.append(km.inertia_)
```

```python

plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```

# Importing libraries

```python
# Import the packages that we will be using
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing data

```python
# url string that hosts our .csv file
path = "datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(path, header=None)

# Assign column names
df.columns = ["sepalLength", "sepalWidth", "petalLength","petalWidth", "class"]
```

```python
# Plot histograms
sns.histplot(df["petalWidth"]).set_title("Petal width histogram")
plt.show()
sns.histplot(df["petalLength"]).set_title("Petal length histogram")
plt.show()
sns.histplot(df["sepalWidth"]).set_title("Sepal width histogram")
plt.show()
sns.histplot(df["sepalLength"]).set_title("Sepal length histogram")
plt.show()
```

```python
# Plot boxplots
sns.boxplot(x=df["petalWidth"]).set_title("Petal width boxplot")
plt.show()
sns.boxplot(x=df["petalLength"]).set_title("Petal length boxplot")
plt.show()
sns.boxplot(x=df["sepalWidth"]).set_title("Sepal width boxplot")
plt.show()
sns.boxplot(x=df["sepalLength"]).set_title("Sepal length boxplot")
plt.show()
```

```python
# Plot petal width boxplots by flower type
sns.boxplot(x=df["petalWidth"], y=df["class"]).set_title("Petal width boxplot by flower class")
plt.show()
```

```python
# Plot sepal length boxplots by flower type
sns.boxplot(x=df["sepalLength"], y=df["class"]).set_title("Sepal length boxplot by flower class")
plt.show()
```

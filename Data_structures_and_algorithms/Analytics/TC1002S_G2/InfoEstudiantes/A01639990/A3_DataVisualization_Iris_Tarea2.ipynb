# Visualizing Data in Python

# Importing libraries

```python
# Import the packages that we will be using
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing data

```python
# url string that hosts our .csv file
ruta = "datasets/iris/iris.csv"



# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(ruta)
#Agregamos los nombres de las columnas
df.columns=["Sepal_length_in_cm","Sepal_width_in_cm", "Petal_length_in_cm", "Petal_width_in_cm", "Class" ] 
```

# Exploring the content of the data set

```python
df.head()
```

# Frequency tables

```python
# Number of times that each distinct value of a variable occurs in a data set
df.Class.value_counts()
```

```python
# Proportion of each distinct value of a variable occurs in a data set
x = df.Class.value_counts()

proporcion = 100 * x/x.sum()

proporcion 
```

```python
# Total number of observations
Nobs,Nvar=df.shape

print(Nobs)


# Total number of null observations
NobsNull=df.Class.isnull().sum()
print(NobsNull)


# Total number of counts (excluding missing values)
NobsNull=df.Class.notnull().sum()
print(NobsNull)
```

# Histogram

```python
# Plot histogram of the Petal width only
X = sns.displot(df["Petal_width_in_cm"])
X.set(ylabel = "Frequency", title = "Histogram of Petal width in cm")


plt.show()
```

```python
# Plot distribution 
sns.histplot(df["Petal_width_in_cm"])
X = sns.histplot(df["Sepal_length_in_cm"])
X.set(ylabel = "Frequency", title = "Histograma de Petal width in cm")


plt.show()
```

# Histograms plotted by groups

```python
# Create histograms of the "Petal width" grouped by "Class"
g= sns.FacetGrid(df,row="Class")
g= g.map(plt.hist, 'Petal_width_in_cm')

plt.show()
```

# Boxplots

```python
# Create the boxplot 
Y = sns.boxplot(df["Petal_width_in_cm"])
Y.set_title("Diagrama de bigotes de Petal width in cm")
plt.show()
```

```python
# Create the boxplot 
sns.boxplot(df["Sepal_length_in_cm"])
Y.set_title("Diagrama de bigotes de Sepal length in cm") 
plt.show()
```

```python
# Create the boxplots of the "Petal width in cm" and of the "Petal length in cm" amounts
sns.boxplot(df["Petal_width_in_cm"])
sns.boxplot(df["Petal_length_in_cm"])

plt.show()

#AVOID THIS KIND OF PLOT

```

# Boxplots plotted by groups

```python
# Create side-by-side boxplots 

Xdata = df.loc[:, ["Petal_width_in_cm","Petal_length_in_cm", "Sepal_width_in_cm", "Sepal_length_in_cm"]]

sns.boxplot(data=Xdata, orient="h")

plt.show()
```

# Histograms and boxplots plotted by groups

```python
# Create a boxplot and histogram

sns.boxplot(y=df["Petal_width_in_cm"], x=df["Class"]).set_title("Boxplot de Petal width in cm de cada clase")

plt.show()
```

# Scatter plot

```python
# scatter plot between two variables
sns.scatterplot(data=df, y ="Petal_width_in_cm", x="Petal_length_in_cm")
plt.show()
```

```python
# scatter plot between two variables (one categorical)
sns.scatterplot(data=df, x="Petal_width_in_cm", y="Class")
plt.show()
```

```python
# scatter plot between two variables (one categorical)
sns.scatterplot(data=df, x ="Sepal_length_in_cm", y="Class")
plt.show()
```

```python
# scatter plot between two variables grouped according to a categorical variable
sns.scatterplot(data=df, y="Sepal_length_in_cm", x ="Sepal_width_in_cm", hue="Class")
plt.show()
```

```python
# scatter plot between two variables grouped according to a categorical variable and with size of markers

sns.scatterplot(data=df, x="Petal_width_in_cm", y="Petal_length_in_cm", hue="Class", size="Sepal_length_in_cm")

plt.show()
```
